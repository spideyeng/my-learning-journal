# git Tips and Tricks

## 1. Create a git repository

```bash
git init
```

## 2. Add files to staging area

```bash
git add <file_name>
```
or add all files to staging area:

```bash
git add .
```

## 3. Commit changes

```bash
git commit -m "<commit_message>"
```

for example:

```bash
git commit -m "Initial commit"
```

## 4. Check status

```bash
git status
```

## 5. Check commit history

```bash
git log
```
