# Python Quick Tips

## Dictionary and DataFrames

###  Convert dictionary to dataframe

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Paris", "London"]
}

df = pd.DataFrame(data)
print(df)
```

### Dictionary of dictionaries

```python
data = {
    "row1": {"Name": "Alice", "Age": 25, "City": "New York"},
    "row2": {"Name": "Bob", "Age": 30, "City": "Paris"},
    "row3": {"Name": "Charlie", "Age": 35, "City": "London"}
}

df = pd.DataFrame.from_dict(data, orient="index")
print(df)
```

### Dictionary with Scalar Values

```python
data = {"Name": "Alice", "Age": 25, "City": "New York"}
df = pd.DataFrame([data])   # wrap in a list
print(df)
```

### Convert Dataframe to Dictionary

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Paris", "London"]
})

dict_default = df.to_dict()
print(dict_default)
```

#### Orient List
Each column becomes a key, values are lists

```python
dict_list = df.to_dict(orient="list")
print(dict_list)
```

#### Orient="records"
Each row becomes a dictionary (great for JSON-like structures):


```python
dict_records = df.to_dict(orient="records")
print(dict_records)
```

#### Orient="index"
Each row index becomes a key, values are dictionaries:


```python
dict_index = df.to_dict(orient="index")
print(dict_index)
```

####  Quick Guide
• 	Use  when exporting to JSON or APIs.
• 	Use  when you want column-wise lists.
• 	Use  when you want row-wise dictionaries.
